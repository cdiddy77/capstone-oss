```python
import json

with open('trello.json', 'r') as file:
    trello_old = json.load(file)

# Print the loaded data
with open('trello2.json', 'r') as file:
    trello_new = json.load(file)

```

```python
foo=[x['id'] for x in trello_old['cards']]
print(foo)
```

```python
old_ids = set([x['id'] for x in trello_old['cards']])
new_ids = set([x['id'] for x in trello_new['cards']])

only_in_old = old_ids - new_ids
only_in_new = new_ids - old_ids

different_date_last_activity = set()

for card_id in old_ids.intersection(new_ids):
    old_card = next((card for card in trello_old['cards'] if card['id'] == card_id), None)
    new_card = next((card for card in trello_new['cards'] if card['id'] == card_id), None)
    if not old_card or not new_card:
        continue
    if old_card['dateLastActivity'] != new_card['dateLastActivity']:
        different_date_last_activity.add(card_id)
```

```python
print('Only in old:', only_in_old)
```

```python
print('Only in new:', only_in_new)
```

```python
print('Different dateLastActivity:', different_date_last_activity)
```

```python
card_details = []

for card_id in different_date_last_activity:
    old_card = next((card for card in trello_old['cards'] if card['id'] == card_id), None)
    new_card = next((card for card in trello_new['cards'] if card['id'] == card_id), None)
    
    if old_card and new_card:
        card_dict = {
            'id': card_id,
            'new_name': new_card['name'],
            'old_name': old_card['name'],
            'new_desc': new_card['desc'],
            'new_labels': new_card['labels'],
        }
        card_details.append(card_dict)

card_details
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
from github import Github
from github import Auth


auth = Auth.Token(os.getenv("GITHUB_TOKEN",''))
g = Github(auth=auth)
```

```python
destination_repo = g.get_repo("diamondkinetics/sbot")

issues_dict = {}

for issue in destination_repo.get_issues():
    issues_dict[issue.title] = issue.number

issues_dict
```

```python

updates = 0

for item in card_details:
    issue_title = item['old_name']
    try:
        issue = destination_repo.get_issue(issues_dict[issue_title])
        print(f"Found issue '{issue_title}' with ID {issue.id}")
    except:
        issue = None
        print(f"Error: Issue '{issue_title}' not found in the repository")
        continue

    issue.edit(title=item['new_name'], body=item['new_desc'])
    print(f"Updated issue {issue.id} : '{issue_title}' to '{item['new_name']}'")
    updates += 1
print(f"Updated {updates} issues")
```
